```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np
```

```python
dataset = pd.read_csv("Salary_Data.csv")
X = dataset["YearsExperience"].values
y= dataset["Salary"].values

```

```python
plt.scatter(X, y)
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.show()
```

```python
#here we are calculating the only the cost function (mean squared error) 
def cost_func(X, y, w, b):
    m = len(X)
    cost_sum = 0
    for i in range(m):
        f = w*X[i]+b
        cost= (f - y[i])**2
        cost_sum +=cost 
    total_cost = (1/(2*m)) *cost_sum
    return total_cost
```

```python
#Calculating the gradients using the mean squared error
def gradient(X, y, w, b):
    m = len(X)
    dw=0
    db= 0
    for i in range(m):
        f= w*X[i]+b
        dw += (f - y[i])
        db += (f-y[i])
    dc_dw = (1/(2*m))* dw
    dc_db = (1/(2*m))* db
    # dc = derivative cost
    return dc_dw, dc_db 
```

```python
#Calculating the gradient descent 
def gradient_descent(X, y, itr, L): #here itr = no. of iterations and L = leanring rate
    # w and b will be the parameters
    w = 0 
    b =0
    for i in range(itr):
        dc_dw , dc_db = gradient(X, y, w, b)
        w = w - L*dc_dw
        b = b - L* dc_db 
        print(f"Iteration {i}: Cost { cost_func(X, y, w, b)}")
    return w, b
```

```python
#let's run the model now 
Learning_rate = 0.01
itr = int(1000)
final_w, final_b = gradient_descent( X, y, itr, Learning_rate)
print(f"Wieght: {final_w:.4f}, Bias: {final_b: .4f}")
```

```python
plt.scatter(X,y, label= "Data Points")
x_vals = np.linspace(min(X), max(X), 100)
y_vals = final_w * x_vals+ final_b
plt.plot(x_vals, y_vals, color = 'r', label =  "Regression Line")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```
